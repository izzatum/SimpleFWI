## Circle reflection

```python
def slowness_model(n):
    # Model
    vback = 2.0
    vel = vback * np.ones(n)
    rr, cc = disk((n[0] // 2, n[1] // 2), radius=20, shape=vel.shape)
    vel[rr, cc] = 2.25

    m = 1. / (vel.reshape(-1, 1)) ** 2
    m0 = 1. / (vback * np.ones(n).reshape(-1, 1)) ** 2

    return m, m0
```

```python
h = np.array([10., 10.])
f = np.array([5., 10., 25.])

xs = np.arange(h[0], 1.e3, 2 * h[0])
zs = 2 * h[0] * np.ones_like(xs)

xr = np.arange(h[0], 1.e3, 2 * h[0])
zr = h[0] * np.ones_like(xr)

n = np.array([91, 101])
z = np.arange(0, n[0]) * h[0]
x = np.arange(0, n[1]) * h[1]
q = np.eye(len(xs))
```

```python
model = {
    'h': h,
    'f': f,
    'xr': xr,
    'zr': zr,
    'xs': xs,
    'zs': zs,
    'n': n,
    'z': z,
    'x': x,
    'q': q}
```

```python
m, m0 = slowness_model(n)

A = getA(5, m, model['h'], model['n'])

plt.figure(figsize=(20, 20))
plt.imshow(np.abs(A.todense())[:9,:9])
plt.axis('tight');
```

```python
m, m0 = slowness_model(n)

plt.figure(figsize=(20, 8))
plt.subplot(121)
plt.imshow(m.reshape(n))
plt.plot(xr // h[0], zr // h[1], 'w^')
plt.plot(xs // h[0], zs // h[1], 'r*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Velocity model', fontsize=15, fontweight='bold')
plt.colorbar()
plt.grid('on', which='both')
plt.axis('tight')

plt.subplot(122)
plt.imshow(m0.reshape(n))
plt.plot(xr // h[0], zr // h[1], 'w^')
plt.plot(xs // h[0], zs // h[1], 'r*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Initial Velocity model', fontsize=15, fontweight='bold')
plt.colorbar()
plt.grid('on', which='both')
plt.axis('tight');
```

```python
Fm = ForwardSolver(model)
U, _ = Fm.solve(m0, rsample=False)

fig, axs = plt.subplots(2, 3, figsize=(20, 11))
axs[0][0].imshow(np.real(U[:, 10, 0]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[0][0].axis('tight')
axs[0][1].imshow(np.real(U[:, 10, 1]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[0][1].axis('tight')
axs[0][2].imshow(np.real(U[:, 10, 2]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[0][2].axis('tight')
axs[1][0].imshow(np.real(U[:, len(zs)//2, 0]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[1][0].axis('tight')
axs[1][1].imshow(np.real(U[:, len(zs)//2, 1]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[1][1].axis('tight')
axs[1][2].imshow(np.real(U[:, len(zs)//2, 2]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[1][2].axis('tight');

U1, _ = Fm.solve(m, rsample=False)

fig, axs = plt.subplots(2, 3, figsize=(20, 11))
axs[0][0].imshow(np.real(U1[:, len(zs)//2, 0]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[0][0].axis('tight')
axs[0][1].imshow(np.real(U1[:, len(zs)//2, 1]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[0][1].axis('tight')
axs[0][2].imshow(np.real(U1[:, len(zs)//2, 2]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[0][2].axis('tight')
axs[1][0].imshow(np.real(U1[:, len(zs)//2, 0]-U[:, len(zs)//2, 0]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[1][0].axis('tight')
axs[1][1].imshow(np.real(U1[:, len(zs)//2, 1]-U[:, len(zs)//2, 1]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[1][1].axis('tight')
axs[1][2].imshow(np.real(U1[:, len(zs)//2, 2]-U[:, len(zs)//2, 2]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[1][2].axis('tight');
```

## Circle transmission

```python
def slowness_model(n):
    # Model
    vback = 2.0
    vel = vback * np.ones(n)
    rr, cc = disk((n[0] // 2, n[1] // 2), radius=20, shape=vel.shape)
    vel[rr, cc] = 2.25

    m = 1. / (vel.reshape(-1, 1)) ** 2
    m0 = 1. / (vback * np.ones(n).reshape(-1, 1)) ** 2

    return m, m0
```

```python
f = np.array([5., 10., 15.])
h = np.array([10., 10.]) # z, x
n = np.array([91, 101]) # z, x

z = np.arange(0, n[0]) * h[0]
x = np.arange(0, n[1]) * h[1]

zs = np.arange(h[0], z[-1], 2 * h[0])
xs = 2 * h[1] * np.ones_like(zs)

zr = np.arange(h[0], z[-1], 2 * h[0])
xr = (n[1]-2) * h[1] * np.ones_like(zr)

q = np.eye(len(xs))
```

```python
model = {
    'h': h,
    'f': f,
    'xr': xr,
    'zr': zr,
    'xs': xs,
    'zs': zs,
    'n': n,
    'z': z,
    'x': x,
    'q': q,
}
```

```python
m, m0 = slowness_model(n)

# plot model
plt.figure(figsize=(20, 8))
plt.subplot(121)
plt.imshow(1./np.sqrt(m).reshape(n), vmin=1.8, vmax=2.2)
plt.plot(xr // h[1], zr // h[0], 'b^')
plt.plot(xs // h[1], zs // h[0], 'r*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Velocity model', fontsize=15, fontweight='bold')
plt.colorbar()
plt.grid('on', which='both')
plt.axis('tight')

plt.subplot(122)
plt.imshow(1./np.sqrt(m0).reshape(n), vmin=1.8, vmax=2.2)
plt.plot(xr // h[1], zr // h[0], 'b^')
plt.plot(xs // h[1], zs // h[0], 'r*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Initial Velocity model', fontsize=15, fontweight='bold')
plt.colorbar()
plt.grid('on', which='both')
plt.axis('tight')
```

```python
Fm = ForwardSolver(model)
U, _ = Fm.solve(m0, rsample=False)

fig, axs = plt.subplots(2, 3, figsize=(20, 11))
axs[0][0].imshow(np.real(U[:, 10, 0]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[0][0].axis('tight')
axs[0][1].imshow(np.real(U[:, 10, 1]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[0][1].axis('tight')
axs[0][2].imshow(np.real(U[:, 10, 2]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[0][2].axis('tight')
axs[1][0].imshow(np.real(U[:, len(zs)//2, 0]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[1][0].axis('tight')
axs[1][1].imshow(np.real(U[:, len(zs)//2, 1]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[1][1].axis('tight')
axs[1][2].imshow(np.real(U[:, len(zs)//2, 2]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[1][2].axis('tight');

U1, _ = Fm.solve(m, rsample=False)

fig, axs = plt.subplots(2, 3, figsize=(20, 11))
axs[0][0].imshow(np.real(U1[:, len(zs)//2, 0]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[0][0].axis('tight')
axs[0][1].imshow(np.real(U1[:, len(zs)//2, 1]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[0][1].axis('tight')
axs[0][2].imshow(np.real(U1[:, len(zs)//2, 2]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[0][2].axis('tight')
axs[1][0].imshow(np.real(U1[:, len(zs)//2, 0]-U[:, len(zs)//2, 0]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[1][0].axis('tight')
axs[1][1].imshow(np.real(U1[:, len(zs)//2, 1]-U[:, len(zs)//2, 1]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[1][1].axis('tight')
axs[1][2].imshow(np.real(U1[:, len(zs)//2, 2]-U[:, len(zs)//2, 2]).reshape(n), cmap='seismic', vmin=-1e2, vmax=1e2)
axs[1][2].axis('tight');
```

## Marmousi

```python
f = loadmat('data/marmousi_50.mat')
vel = f['vel']

# smooth velocity model
vel0 = gaussian_filter(vel, sigma=15)
```

```python
# spatial axes
dx, dz = 10, 10
h = np.array([dz, dx])
n = np.array(vel.shape)

z = np.arange(0, n[0]) * dz
x = np.arange(0, n[1]) * dx

# frequencies
f = np.array([5., 10., 25.])

# sources position
xs = np.arange(dx, n[1]*dx, 5 * dx)
zs = 4 * dx * np.ones_like(xs)

# receivers position
xr = np.arange(dx, n[1]*dx, 3 * dx)
zr = 2 * dx * np.ones_like(xr)

# sources matrix
q = np.eye(len(xs))
```

```python
model = {
    'h': h,
    'f': f,
    'xr': xr,
    'zr': zr,
    'xs': xs,
    'zs': zs,
    'n': n,
    'z': z,
    'x': x,
    'q': q,
}
```

```python
m, m0 = 1./(vel.reshape(-1, 1)) ** 2, 1./(vel0.reshape(-1, 1)) ** 2

plt.figure(figsize=(20, 8))
plt.subplot(121)
plt.imshow(m.reshape(n))
plt.plot(xr // h[0], zr // h[1], 'w^')
plt.plot(xs // h[0], zs // h[1], 'r*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Velocity model', fontsize=15, fontweight='bold')
plt.colorbar()
plt.grid('on', which='both')
plt.axis('tight')

plt.subplot(122)
plt.imshow(m0.reshape(n))
plt.plot(xr // h[0], zr // h[1], 'w^')
plt.plot(xs // h[0], zs // h[1], 'r*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Initial Velocity model', fontsize=15, fontweight='bold')
plt.colorbar()
plt.grid('on', which='both')
plt.axis('tight');
```

```python
Fm = ForwardSolver(model)
U, _ = Fm.solve(m0.T, rsample=False)

clip = 1e2

fig, axs = plt.subplots(2, 3, figsize=(20, 4))
axs[0][0].imshow(np.real(U[:, 10, 0]).reshape(n), cmap='seismic', vmin=-clip, vmax=clip)
axs[0][0].axis('tight')
axs[0][1].imshow(np.real(U[:, 10, 1]).reshape(n), cmap='seismic', vmin=-clip, vmax=clip)
axs[0][1].axis('tight')
axs[0][2].imshow(np.real(U[:, 10, 2]).reshape(n), cmap='seismic', vmin=-clip, vmax=clip)
axs[0][2].axis('tight')
axs[1][0].imshow(np.real(U[:, len(zs)//2, 0]).reshape(n), cmap='seismic', vmin=-clip, vmax=clip)
axs[1][0].axis('tight')
axs[1][1].imshow(np.real(U[:, len(zs)//2, 1]).reshape(n), cmap='seismic', vmin=-clip, vmax=clip)
axs[1][1].axis('tight')
axs[1][2].imshow(np.real(U[:, len(zs)//2, 2]).reshape(n), cmap='seismic', vmin=-clip, vmax=clip)
axs[1][2].axis('tight');
```

## Dot-test on Jacobian

```python
def slowness_model(n):
    # Model
    vback = 2.0
    vel = vback * np.ones(n)
    rr, cc = disk((n[0] // 2, n[1] // 2), radius=20, shape=vel.shape)
    vel[rr, cc] = 2.25

    m = 1. / (vel.reshape(-1, 1)) ** 2
    m0 = 1. / (vback * np.ones(n).reshape(-1, 1)) ** 2

    return m, m0
```

```python
f = np.array([5., 10., 15.])
h = np.array([10., 10.]) # z, x
n = np.array([91, 101]) # z, x

z = np.arange(0, n[0]) * h[0]
x = np.arange(0, n[1]) * h[1]

zs = np.arange(h[0], z[-1], 2 * h[0])
xs = 2 * h[1] * np.ones_like(zs)

zr = np.arange(h[0], z[-1], 2 * h[0])
xr = (n[1]-2) * h[1] * np.ones_like(zr)

q = np.eye(len(xs))
```

```python
model = {
    'h': h,
    'f': f,
    'xr': xr,
    'zr': zr,
    'xs': xs,
    'zs': zs,
    'n': n,
    'z': z,
    'x': x,
    'q': q,
}
```

```python
m, m0 = slowness_model(n)

# plot model
plt.figure(figsize=(20, 8))
plt.subplot(121)
plt.imshow(1./np.sqrt(m).reshape(n), vmin=1.8, vmax=2.2)
plt.plot(xr // h[1], zr // h[0], 'b^')
plt.plot(xs // h[1], zs // h[0], 'r*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Velocity model', fontsize=15, fontweight='bold')
plt.colorbar()
plt.grid('on', which='both')
plt.axis('tight')

plt.subplot(122)
plt.imshow(1./np.sqrt(m0).reshape(n), vmin=1.8, vmax=2.2)
plt.plot(xr // h[1], zr // h[0], 'b^')
plt.plot(xs // h[1], zs // h[0], 'r*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Initial Velocity model', fontsize=15, fontweight='bold')
plt.colorbar()
plt.grid('on', which='both')
plt.axis('tight');
```

```python
# Dot-test with Dobs and m0
Fm = ForwardSolver(model)
Dobs, DF = Fm.solve(m0)

(DF * m0).conj().T @ Dobs, np.conj(m0).T @ (DF.H * Dobs)
```

```python
# Dot-test with complex vectors
n1 = len(xr)*len(xs)*len(f)
n2 = 91*101

v1 = 2.5*np.ones(n1) + 2.j*np.ones(n1)
v2 = 3.5*np.ones(n2) - 1.75j*np.ones(n2)

(DF * v2).conj().T @ v1, np.conj(v2).T @ (DF.H * v1)
```
